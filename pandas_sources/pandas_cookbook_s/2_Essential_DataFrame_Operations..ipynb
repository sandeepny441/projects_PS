```python
df_college_ugds = df_college.filter(like='UGDS_')
```

```python
df_college_ugds.head()
```

```python
df_college_ugds.isnull().sum(axis =1).sort_values(ascending = False).head()
```

```python
df_college_ugds = df_college_ugds.dropna(how = 'all')
df_college_ugds.isnull().sum()
```

```python
df_college_ugds.ge(0.15).head()
```

```python
diversity_metric = df_college_ugds.ge(.15).sum(axis = 'columns')
```

```python
diversity_metric.head()
```

```python
diversity_metric.value_counts()
```

```python
diversity_metric.sort_values(ascending=False).head()
```
